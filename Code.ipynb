```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
```

# Data Cleaning

## Whale Returns

```python
# Reading whale_returns csv file and convering it into data frame
whale_returns= pd.read_csv(Path("whale_returns.csv"), index_col='Date', infer_datetime_format=True, parse_dates=True)

whale_returns.sort_index()
whale_returns.head()
whale_returns.dtypes
```

```python
# counting nulls
whale_returns.isnull().sum()
```

```python
# dropping nulls
whale_returns=whale_returns.dropna()
whale_returns.isnull().sum()
```

```python
whale_returns.head()
```

```python
whale_returns.dtypes
```

## Algorithmic daily Returns

```python
algo_returns=pd.read_csv(Path("algo_returns.csv"), index_col='Date', infer_datetime_format=True, parse_dates=True)
algo_returns=algo_returns.sort_index()
```

```python
# counting nulls
algo_returns.isnull().sum()
```

```python
# dropping nulls
algo_returns=algo_returns.dropna()
algo_returns.isnull().sum()
```

```python
algo_returns.head()
```

## S&P 500 Returns

```python
# Reading S&P500 csv file, and sorting the dates
sp500_history=pd.read_csv(Path("sp500_history.csv"), index_col='Date', infer_datetime_format=True, parse_dates=True)
sp500_history=sp500_history.sort_index(ascending=True)
sp500_history.head()
```

```python
sp500_history.dtypes
```

```python
# changing the Close column to float data type by removing the string symbols $
sp500_history['Close']=sp500_history['Close'].str.replace('$', "")
#sp500_history['Close']=sp500_history['Close'].str.replace(',', "")
sp500_history['Close']=sp500_history['Close'].astype("float")

sp500_history.head()
```

```python
#checking datatype
sp500_history.dtypes
```

```python
#Calculating the S&P500 returns
sp500_history['S&PReturns']=sp500_history.pct_change()
sp500_history.head()
```

```python
#droppping null values
sp500_history=sp500_history.dropna()
sp500_history.head()
```

```python
#dropping the "Close" column to create a dataframe containing only the S&P500 returns
sp500_returns=sp500_history.drop(columns=['Close'])
sp500_returns.head()
```

## Combining the Whale, Algorithmic and S&P500 dataframe

```python
portfolio=pd.concat([whale_returns,algo_returns,sp500_returns],axis='columns', join='inner')
portfolio.head()
```

## Quantitaive Analysis

### Performance Analysis

#### Plotting Daily Returns

```python
#plotting daily returns of the portfolio
portfolio.plot(figsize=(15,15),title="Daily returns")
```

#### Plotting cumulative returns

```python
cumulative_returns=(1+portfolio).cumprod()
cumulative_returns.plot(figsize=(15,15),title='Cumulative_returns plot')
```

## Risk Analysis

### Box plot of the Portfolio

```python
portfolio.plot.box(figsize=(20,10), title="Portfolio Risk")
```

 ### Standard Deviation of funds in the Portfolio

```python
portfolio.std()
```

### Determining Portfolios riskier than S&P500

```python
#Calculating risk of S&P500 and cmparing it with the risk of other funds
sp500_risk=portfolio["S&PReturns"].std()    
portfolio.std()>sp500_risk               
```

### Annualised Standard Deviation

```python
annualised_std=portfolio.std()*np.sqrt(252)
annualised_std
```

### Rolling Statistics

```python
portfolio_roll_std=portfolio.rolling(window=21).std()
portfolio_roll_std.plot(figsize=(20,10),title="21 Day rolling standard deviation")
```

### Corelation and its plot

```python
cor_portfolio= portfolio.corr()
cor_portfolio
```

### Calculating and plotting the beta for Algo 1 and S&P500

```python
covar=portfolio['Algo 1'].rolling(window=20).cov(portfolio['S&PReturns'])
var=portfolio['S&PReturns'].rolling(window=21).var()
beta=covar/var
beta.plot(figsize=(15,15), title="Algo1 Beta")
```

### Exponentially Weighted Standard deviation

```python
expo_weighted=portfolio.ewm(halflife=21).std()
#print(expo_weighted)
expo_weighted.plot(figsize=(15,15), title="Exponentially weighted Standard Deviation")
```

### Sharpe Ratios

```python
sharpe_ratio=portfolio.mean()*252/(portfolio.std()*np.sqrt(252))
sharpe_ratio
```

```python
#Plotting Sharpe ratios
sharpe_ratio.plot.bar()
```

# The Algo1 strategy outperforms S&P500 and the Whales Portfolio, because it has the Highest Sharpe Ratio, but we have to be cautious, since this also indicates that it is the riskiest

# Custom Portfolio

## Reading data of the three Stocks

```python
# Reading data of Google
goog_historical= pd.read_csv(Path("goog_historical.csv"), index_col='Trade DATE', infer_datetime_format=True,parse_dates=True)
goog_historical=goog_historical.sort_index()
goog_historical.head()
```

```python
# Reading data of Apple
aapl_historical= pd.read_csv(Path("aapl_historical.csv"), index_col='Trade DATE', infer_datetime_format=True,parse_dates=True)
aapl_historical=aapl_historical.sort_index()
aapl_historical.head()
```

```python
#Reading data for Costco
cost_historical=pd.read_csv(Path("cost_historical.csv"), index_col='Trade DATE', infer_datetime_format=True,parse_dates=True)
cost_historical=cost_historical.sort_index()
cost_historical.head()
```

## Combining each dataframe to create a Portfolio

## Calculating Weighted Returns of the Stocks assuming an equal numbr of Shares of each stock

```python
custom_portfolio=pd.concat([goog_historical, aapl_historical, cost_historical], axis='rows',join='inner')
custom_portfolio=custom_portfolio.reset_index()
custom_portfolio=custom_portfolio.pivot_table(values="NOCP",index="Trade DATE", columns="Symbol")
custom_portfolio.head()
```

### Daily Returns dataframe and Calculating Weighted Daily returns

```python
daily_returns=custom_portfolio.pct_change()
daily_returns.dropna().head()
```

```python
#Setting equal weights
weights=[1/3,1/3, 1/3]
custom_portfolio_weighted_returns = daily_returns.dot(weights)
custom_portfolio_weighted_returns
```

```python
#joining weighted return to the daily returns of individual stock
daily_returns['weightedreturn']=custom_portfolio_weighted_returns
daily_returns.head()
```

```python
portfolio['Customreturns']= custom_portfolio_weighted_returns 
portfolio.tail()
```

## Risk Analysis

```python
#Craeting a Dataframe by dropping NA values, so that we can run risk analysis for the portfolio
portfolio=portfolio.dropna()
portfolio.head()
```

### Calculating Annualised Risk

```python
portfolio_risk=portfolio.std()*np.sqrt(252)
portfolio_risk
```

### Calculating and Plotting Rolling Statistics

```python
portfolio_roll_risk=portfolio.rolling(window=21).std()
portfolio_roll_risk=portfolio_roll_risk.dropna()
portfolio_roll_risk.head()
```

```python
portfolio_roll_risk.plot(figsize=(21,10), title='21 Day Rolling Risk')
```

## Calculating and Plotting Corelation

```python
cor= portfolio.corr()
cor
```

## Calculating and Plotting 60 day rolling Beta for custom Portfolio and S&P 500

```python
cov=portfolio['Customreturns'].rolling(window=60).cov(portfolio['S&PReturns'])
var=portfolio['S&PReturns'].rolling(window=60).var()
beta=cov/var
beta.plot(figsize=(20,10), title="Beta plot for Custom Portfolio")
```

## Sharpe Ratios

```python
sharpe_ratios= (portfolio.mean()*252)/(portfolio.std()*np.sqrt(252))
sharpe_ratios
```

```python
sharpe_ratios.plot.bar()
```

# The Custom Portfolio does good. It has lower sharpe ratio than Algo 1, but it is also less riskier than Algo 1. I think its the best performer after Algo 1
